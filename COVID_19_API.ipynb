```python
import requests

url = "https://covid-19-india-data-by-zt.p.rapidapi.com/GetIndiaDistrictWiseDataForState"

querystring = {"statecode":"GJ"}

headers = {
    'x-rapidapi-host': "covid-19-india-data-by-zt.p.rapidapi.com",
    'x-rapidapi-key': "a48f66be4dmsh6b9b5887f57da50p160e46jsn082500ddf7bd"
    }

response = requests.request("GET", url, headers=headers, params=querystring)

#print(response.text)
```

```python
sk = response.json()
```

```python
sk
```

```python
sk['data'][0]
```

```python
Dist = []
for var in range(len(sk['data'])):
    Dist.append(sk['data'][var]['name'])
    
print(Dist)    
```

```python
Active_Case = []
for var in range(len(sk['data'])):
    Active_Case.append(sk['data'][var]['active'])
    
print(Active_Case)
```

```python
Death_Case = []
for var in range(len(sk['data'])):
    Death_Case.append(sk['data'][var]['deceased'])
    
print(Death_Case)
```

```python
Conf_Case = []
for var in range(len(sk['data'])):
    Conf_Case.append(sk['data'][var]['confirmed'])
    
print(Conf_Case)
```

```python
Discharge_Case = []
for var in range(len(sk['data'])):
    Discharge_Case.append(sk['data'][var]['recovered'])
    
print(Discharge_Case)
```

```python
Nconf_Case = []
for var in range(len(sk['data'])):
    Nconf_Case.append(sk['data'][var]['newconfirmed'])
    
print(Nconf_Case)
```
